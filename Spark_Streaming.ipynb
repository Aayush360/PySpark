```
from pyspark.streaming import StreamingContext
from pyspark import SparkConf,SparkContext



conf = SparkConf().setAppName('Streaming')
sc = SparkContext.getOrCreate(conf=conf)

ssc = StreamingContext(sc,1) #duration after which it will look for the directory, if there is new file it will simply take it on 
# ssc = StreamingContext.getActiveOrCreate(sc,1)
```

```
rdd=ssc.textFileStream("/FileStore/tables/") # start reading the data that is present inside the directory
# this is not the regular rdd but a datastream
```

```
# spark streaming with RDD
```

```
rdd = rdd.map(lambda x:(x,1))
rdd = rdd.reduceByKey(lambda x,y: x+y)
rdd.pprint() # prettyprint




ssc.start() # will start reading the data from the directory and keeps looking on it

# only considers those files that comes into the directory after starting the session, if file is already present it will simply ignore it

ssc.awaitTerminationOrTimeout(100)
# if either user terminates manually or for 100 seconds there is no file, it will simply terminate it
```
