```
from pyspark import SparkConf, SparkContext
```

```
conf = SparkConf().setAppName("Read File")
```

```
sc=SparkContext.getOrCreate(conf=conf)
```

```
text = sc.textFile('/FileStore/tables/hello.txt') # text is an rdd
```

```
text.collect()
```

```
## MAP
```

```
new_rdd=text.map(lambda x: x.split())
```

```
new_rdd.collect()
```

```
new_rdd=text.map(lambda x: x+' aayush')
```

```
new_rdd.collect()
```

```
def foo(x):
  return x.split(' ')

rdd_new = new_rdd.map(foo)
rdd_new.collect()
```

```
def adder(x):
  l = x.split()
  li=[]
  for x in l:
    li.append(int(x)+10)
  return li

rdd_added = text.map(adder)
rdd_added.collect()
```

```
def filter_words(x):
  a = x.split(' ')
  return a
          
    
  
```

```
fil_rdd = test_temp.filter(filter_words)
fil_rdd.collect()
```

```
fil_r = test_temp.flatMap(filter_words)
fil_r.collect()
```

```
def filter_a_c(x):
  if x.startswith('a') or x.startswith('c'):
    return False
  else:
    return True
  
```

```
test_temp.map(lambda x: (x,len(x.split(' ')))).collect()
```

```
test_temp.flatMap(lambda x: x.split(' ')).map(lambda x:(x,len(x))).groupByKey().mapValues(list).collect()
```

```
test_temp.flatMap(lambda x: x.split(' ')).map(lambda x:(x,len(x))).reduceByKey(lambda x,y:x+y).collect() # we can see that ant value has been added to 3+3=6
```
